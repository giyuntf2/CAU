# 3차 함수 $f(x) = (x-2)x(x+2) 그리기$



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x-2) * x * (x+2)

x = np.arange(-3, 3, 0.5)
y = f(x)

plt.plot(x, y)
plt.show()
```

## 3.1.6. 그래프를 장식하기


```python
def f(x, w):
    return (x-w) * x * (x+2)

x = np.linspace(-3, 3, 100)

plt.plot(x, f(x, 1), color='blue', label='$w=1$')
plt.plot(x, f(x, 2), color='black', label='$w=2$')
plt.plot(x, f(x, 3), color='red', label='$w=3$')
plt.legend(loc="upper left")
plt.grid(True)
plt.ylim(-15, 15)
plt.show()
```

## 3.1.7. 그래프를 여러개 보여주기
plt.subplot(n1, n2, n)을 사용하면 세로 n1, 가로 n2로 나눈 n번째 그래프가 그려진다.


```python
plt.figure(figsize=(10, 3))
plt.subplots_adjust(wspace=0.5, hspace=0.5) #w가 좌우 간격, h가 상하간격

for i in range(0, 6):
    plt.subplot(2, 3, i+1)
    plt.title(i + 1)
    plt.plot(x, f(x, i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

# 3.2 3차원 그래프 그리기
## 3.2.1 이변수 함수

$f(x_0, x_1) = (2x^2_0 + x^2_1) * exp(-(2x^2_0 + x^2_1))$

```python
def f3(x_0, x_1):
    r = 2 * x_0 ** 2 + x_1 ** 2
    ans = r * np.exp(-r)

    return ans

x_0 = np.linspace(-2, 2, 100)
x_1 = np.linspace(-2, 2, 100)

y = np.zeros((len(x_0), len(x_1)))

for i0 in range(100):
    for i1 in range(100):
        y[i1, i0] = f3(x_0[i0], x_1[i1])

np.round(y, 1)
```

```python
plt.bone()
plt.pcolor(y)
plt.colorbar()
plt.show()
```

## 3.2.3. 함수의 표면을 표시 : Surface

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x_0, x_1) #그리드를 만들때 격자에 점을 찍어줌.

plt.scatter(xx0, xx1)
plt.grid(True)
```

```python
#기본 surface

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1,1,1, projection='3d')
ax.plot_surface(xx0, xx1, y, color='blue', edgecolor='black')
plt.show()
```

```python
def get_3dsurface(rstride, cstride): #rstride, cstride의 숫자는 자연수. 숫자가 작을수록 조밀해짐
    global xx0, xx1
    plt.figure(figsize=(5, 3.5))
    ax = plt.subplot(1,1,1, projection='3d')
    ax.plot_surface(xx0, xx1, y, rstride = rstride, cstride = cstride,
                    alpha=0.8, color='blue', edgecolor='black')
    ax.set_zticks((0, 0.2))
    ax.view_init(75, -95) #각도 조절 가능
    plt.show()

get_3dsurface(1, 1)
```

```python
def get_3dsurface1(rstride, cstride): #rstride, cstride의 숫자는 자연수. 숫자가 작을수록 조밀해짐
    global xx0, xx1
    plt.figure(figsize=(5, 3.5))
    ax = plt.subplot(1,1,1, projection='3d')
    ax.plot_surface(xx0, xx1, y, rstride = rstride, cstride = cstride,
                    alpha=0.8, color='blue', edgecolor='black')
    ax.set_zticks((0, 0.2))
    ax.view_init(90, -95) #각도 조절 가능
    plt.show()

get_3dsurface1(1, 1)
```

## 3.2.4 등고선으로 표시 : contour

```python
#%reset

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def f3(x_0, x_1):
    r = 2 * x_0 ** 2 + x_1 ** 2
    ans = r * np.exp(-r)

    return ans

x0 = np.linspace(-2, 2, 50)
x1 = np.linspace(-2, 2, 50)

y = np.zeros((len(x0), len(x1)))

for i in range(len(x0)):
    for j in range(len(x1)):
        y[i, j] = f3(x0[j], x1[i])

xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, color='black')
cont.clabel(fmt='%3.2f', fontsize=8)
plt.xlabel('%x_0%', fontsize=14)
plt.ylabel('%x_1%', fontsize=14)
plt.show()
```
