# Matrix Dot Product


$Projection Vector = \displaystyle\frac{a^T*b}{\sqrt{b^T * b}} \frac{b}{\sqrt{b^T * b}}$

```python
%reset
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
v = np.random.randint(1, 10, size=2)
w = np.random.randint(1, 10, size=2)
print(v, w)

def get_proj_to_w(v, w):
    return (v.T * w / np.sqrt(w.T * w)) * (w / np.sqrt(w.T * w))

def get_proj_to_v(v, w):
    return (w.T * v / np.sqrt(v.T * v)) * (v / np.sqrt(v.T * v))

get_proj(v, w).dot(w)
```

```python
import math
def dist(v):
    return math.sqrt(v[0]**2 + v[1]**2)
```

```python
def innerproduct(v, w):
    distA = dist(v)
    distB = dist(w)
    
    print(distA, distB)

    #내적
    ip = v[0] * w[0] + v[1] * w[1]

    ip_2 = distA * distB

    cost = ip / ip_2

    print("cos x : %10.3f" % cost)

    x = math.acos(cost)

    print("x (radians) : %10.3f" % x)

    degree = math.degrees(x)

    print("x (degrees) : %10.3f" % degree)

    return degree

innerproduct(v, w)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,10))
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(ticks=np.arange(0, 10, step=1))
plt.yticks(ticks=np.arange(0, 10, step=1))
plt.grid(True)
plt.arrow(0, 0, v[0], v[1], head_width = .1, head_length = .1, color = 'red')
plt.arrow(0, 0, w[0], w[1], head_width = .1, head_length = .1, color = 'green')
plt.show()
```

```python
def get_proj(a, b):
    return (a.T * b / np.sqrt(b.T * b)) * (b / np.sqrt(b.T * b))

a = np.array([6,9])
b = np.array([6,1])
```

```python
np.dot(a, b)
```

```python
# 내적 1
ip = get_proj(a, b).dot(b)

# 내적 2
distA = np.linalg.norm(a)
distB = np.linalg.norm(b)

ip_2 = distA * distB

ip, ip_2

cost_1 = ip / ip_2

print("cos x : %10.3f" % cost_1)

x_ = math.acos(cost_1)

print("x (radians) : %10.3f" % x_)

degrees = math.degrees(x_)

print("x (degrees) : %10.3f" % degrees)
```
