```python
np.random.seed(1)
v = np.random.randint(1, 10, size=2)
w = np.random.randint(1, 10, size=2)
print(v, w)

def get_proj_to_w(v, w):
    return (v.T * w / np.sqrt(w.T * w)) * (w / np.sqrt(w.T * w))

def get_proj_to_v(v, w):
    return (w.T * v / np.sqrt(v.T * v)) * (v / np.sqrt(v.T * v))

get_proj(v, w).dot(w)
```

```python
import math
def dist(v):
    return math.sqrt(v[0]**2 + v[1]**2)
```

```python
def innerproduct(v, w):
    distA = dist(v)
    distB = dist(w)
    
    print(distA, distB)

    #내적
    ip = v[0] * w[0] + v[1] * w[1]

    ip_2 = distA * distB

    cost = ip / ip_2

    print("cos x : %10.3f" % cost)

    x = math.acos(cost)

    print("x (radians) : %10.3f" % x)

    degree = math.degrees(x)

    print("x (degrees) : %10.3f" % degree)

    return degree

innerproduct(v, w)
```

왼쪽은 '행렬 표기(벡터 표기)', 오른쪽은 '성분 표기' 라고 부른다.

4.4 미분

머신러닝은 결국 함수에서 최소나 최대인 입력을 찾는 최적화 문제이다. 함수의 최소 지점은 기울기가 0이 되는 성질이 있으므로, 이러한 문제를 풀려면 함수의 기울기를 아는 것이 중요하다. 그 함수의 기울기를 도출하는 방법이 '미분'이다.



4.4.1 다항식의 미분

$f(w) = w^2$

power rule을 통해서 미분이 가능하다.

$\displaystyle power \ rule : \frac{d}{dw}w^n = nw^{n-1} $

$\displaystyle \frac{dw^2}{dw} = 2w$

image.png

## 4.4.2 미분 기호가 들어간 수식의 변형

미분의 기호가 들어간 수식의 변형을 생각해보자. 미분 기호 $\frac{d}{dw}$는 오른쪽에만 작용한다. 합의 기호 $\sum$도 마찬가지이다. $2w^5$와 같이, 숫자가 $w^n$의 앞에서 곱해진 경우, 그 숫자는 미분 기호 왼쪽에 나타낼 수 있다.

미분과 무관한 부분은 문자식이라도 왼쪽에 나타낼 수는 있다. 하지만, 그 값이 포함되어 있지 않으면 0으로 처리 된다.

##4.4.3 중첩된 함수의 미분
합성함수 미분법을 생각하자.

$f(w) = g(w)^2\ (식 4.31) \\ g(w) = aw+b\ (식 4.32)$

이는 간단하게 다음과 같이 미분이 가능하다.

$\displaystyle\frac{df(w)}{dw}=a^2w^2+2abw+b^2 \\
= 2a^2w + 2ab$

##4.4.4 중첩된 함수의 미분 : 연쇄법칙

$\displaystyle\frac{df}{dw} = \frac{df}{dg} \cdot \frac{dg}{dw}$

먼저 df/dg 부분은 'f를 g로 미분한다' 라는 의미이며, 미분 공식을 적용하면 다음과 같다.

$\displaystyle\frac{df}{dg} = \frac{dg^2}{dg} = 2g$

그리고 dg/dw는 g를 w로 미분한다는 의미이며 다음과 같다.

$\displaystyle\frac{dg}{dw} = \frac{d}{dw}(aw+b) = a$

따라서, 이것에 대해 chain rule을 사용하게 되면 다음과 같은 결과를 얻을 수 있다.

$\displaystyle\frac{df}{dw} = \frac{df}{dg} \cdot \frac{dg}{dw} = 2ga = 2(aw+b)a = 2a^2w + 2ab$

#4.5 편미분

##4.5.1 편미분이란?

머신러닝에서 실제로 사용하는 것은 순수한 미분이 아닌, 편미분이다.

이변수 함수를 예로 들어보겠다.

$f(w_0, w_1) = w^2+2w_0w_1+3$

image.png

## 4.5.2 편미분과 도형

image.png

### w0에서 자르기
$\displaystyle \frac{\partial{f}}{\partial{w_0}}$를 이해하려면 w0 축에 평행이 되도록 f를 칼로 잘랐을 때의 단면을 생각해보자. (그림 4-12 1.) 

단면은 아래로 볼록한 2차 함수이다. 이 단면의 곡선의 기울기가 

$\displaystyle \frac{\partial{f}}{\partial{w_0}}=2w_0+2w_1$ 로 되어있다. (그림 4-12 2.)

w_1 = -1의 면을 자르면 미분하기 전 식에 w1 = -1을 대입한 식이 이 면에서 기울기의 식이 된다.

$\displaystyle \frac{\partial{f}}{\partial{w_0}}|_{w_1=-1}$ 처럼 표현이 가능하다.

$\displaystyle \frac{\partial{f}}{\partial{w_0}}|_{w_1=-1} = 2w_0+2w_1|_{w1=-1} = 2w_0-2$

w0의 축에 평행이 되는 면은 수없이 많다. 

예를 들어 w1 = 1의 면에서 자르면 f의 단면은 [그림 4-12 3.]과 같이 되어, 단면의 기울기는 아래와 같이 된다.

$\displaystyle \frac{\partial{f}}{\partial{w_0}}|_{w_1=-1} = 2w_0+2w_1|_{w1=1} = 2w_0+2$



### w1에서 자르기
한편 $\displaystyle \frac{\partial{f}}{\partial{w_1}}$은 w1축에 평행이 되도록 f를 끌 때 단면이다.

이 단면은 직선이다. 그림 5번 참조.

w0 = -1의 면에서 자른 단면의 기울기는 아래와 같이 된다.

$\displaystyle \frac{\partial{f}}{\partial{w_1}}|_{w_1=1} = 2w_0|_{w0=1}=2$

w0 = 1의 면에서 자른 단면의 기울기는 아래와 같이 된다.

$\displaystyle \frac{\partial{f}}{\partial{w_1}}|_{w_1=1} = 2w_0|_{w0=-1}=-2$

**이 두 개의 기울기를 세트로 하여 벡터로 해석할 수 있다.**

이것을 f의 w에 대한 경사 (기울기 벡터, gradient)라고 부르며, 아래와 같이 나타낸다.

$\displaystyle ∇_wf=\begin{bmatrix}
\frac{\partial{f}}{\partial{w_0}}\\
\frac{\partial{f}}{\partial{w_1}}
\end{bmatrix}$

경사는 기울기가 가장 큰 방향과 그 크기를 나타낸다.

```python
from sympy import *
import sympy as sy
from sympy.plotting import plot3d

x, y = sy.symbols('x y')
plot3d(x**2 + 2*x*y + 3)
```

```python
w0, w1 = symbols('w0 w1')
f = w0**2 + 2*w0*w1 + 3
df_w0 = sy.diff(f,w0)
df_w1 = sy.diff(f,w1)

print(df_w0, df_w1)
```

```python
plot3d(f, df_w0.subs({w1:-1}))
```

```python
#그림 5,6번 표현 
#f(w0, w1)을 w0 = 1로 자른 단면
#그 기울기가 w1에 관한 편미분(w0=1 일 대)
plot(df_w1, df_w1.subs({w0:1}))
```

```python
#그림 7번 표현
#f(w0, w1)을 w0 = -1로 자른 단면
#그 기울기가 w1에 관한 편미분(w0=1 일 대)
plot(df_w1, df_w1.subs({w0:-1}))
```

###4.5.3 경사를 그림으로 나타내기


```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1):
    return w0**2 + 2 * w0 * w1 + 3

def df_dw0(w0, w1):
    return 2 * w0 + 2 * w1

def df_dw1(w0, w1):
    return 2 * w0 + 0 * w1

w_range = 2
dw = 0.25

w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]

ww0, ww1 = np.meshgrid(w0, w1)

ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros((len(w0), len(w1)))
dff_dw1 = np.zeros((len(w0), len(w1)))

for i0 in range(wn):
    for i1 in range(wn):
        ff[i1, i0] = f(w0[i0], w1[i1])
        dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
        dff_dw1[i1,x i0] = df_dw1(w0[i0], w1[i1])
        
plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace= 0.3)
plt.subplot(1, 2, 1)
cont = plt.contour(ww0, ww1, ff, 10, colors='k') # fの等高線表示
cont.clabel(fmt='%2.0f', fontsize = 8)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - .5, w_range + .5)
plt.ylim(-w_range - .5, w_range + .5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

plt.subplot(1, 2, 2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1) # fの勾配ベクトル表示
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range - .5, w_range + .5)
plt.ylim(-w_range - .5, w_range + .5)
plt.show()
```

## 4.6.8 행렬과 사상

행렬은 '벡터를 다른 벡터로 변환하는 규칙'으로 해석할 수 있다.

벡터라는 숫자의 조합을 '좌표'라고 해석하면 그것은 '공간에 있는 점' 으로도 볼 수 있기 때문에 '어떤 점을 다른 점으로 이동시키는 규칙'으로 파악할 수 있다.

이처럼 그룹 (벡터와 점)에서 그룹 (벡터와 점)에 대응 관계를 제공하는 규칙을 사상이라고 하며, 행렬은 선형 사상으로 분류된다.



```python
import numpy as np
import matplotlib.pyplot as plt

XMIN, XMAX, YMIN, YMAX = -3, 3, -3, 3
N = 16
xgrid = np.linspace(XMIN, XMAX, N)
ygrid = np.linspace(XMIN, XMAX, N)
grid = np.array(np.meshgrid(xgrid, ygrid)).reshape(2, N**2)

# standard basis
basis = np.array([[1,0], [0,1]])

def plot_quad(basis, color='k'):
    ix, iy = basis[0]
    jx, jy = basis[1]
    plt.plot([0, ix, ix+jx, jx, 0], [0, iy, iy+jy, jy, 0], color)

def plot_vector(v, color='k', lw=1):
    plt.plot([0, v[0]], [0, v[1]], c=color, lw=lw)

def plot_points(grid, color='k'):
    plt.scatter(*grid, c=color, s=2, alpha=0.5)

def apply_transformation(basis, T):
    return (T @ basis.T).T

# The untransformed grid and unit square.
plot_points(grid)
plot_quad(basis)

# Apply the transformation matrix, S, to the scene.
S = np.array(((1.5, 0.5),(0.5, 1.5)))
tbasis = apply_transformation(basis, S)
plot_quad(tbasis, 'r')
tgrid = S @ grid
plot_points(tgrid, 'r')

# Find the eigenvalues and eigenvectors of S... 
vals, vecs = np.linalg.eig(S)
print(vals, vecs)
if all(np.isreal(vals)):
    # ... if they're all real, indicate them on the diagram.
    v1, v2 = vals
    e1, e2 = vecs.T
    plot_vector(v1*e1, 'r', 3)
    plot_vector(v2*e2, 'r', 3)
    plot_vector(e1, 'k')
    plot_vector(e2, 'k')

# Ensure the plot has 1:1 aspect (i.e. squares look square) and set the limits.
plt.axis('square')
plt.xlim(XMIN, XMAX)
plt.ylim(YMIN, YMAX)

plt.show()
```

## 4.7.8 가우스 함수

$y = exp(-x^2)$

x = 0 을 중심으로 종모양을 하고 있다.

$\displaystyle y = aexp(-\frac{(x-\mu)^2}{σ^2})$

이때, $\mu$ : 중심 (평균), σ : 확산 (표준 편차), $\alpha$ : 높이

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)

def gaussian_function(a, mu, sigma, x):
    y = np.exp(-x**2)
    y1 = a * np.exp(-(x-mu)**2 / sigma**2)

    plt.figure(figsize=(4, 4))
    plt.plot(x, y, 'black', linewidth=3, label='$y=exp(-x^2)$')
    plt.plot(x, y1, 'red', linewidth=3, label='$y=0.5exp(-\frac{(x-2)^2}/{3^2})$')
    plt.grid(True)
    plt.xlim(-4, 4)
    plt.ylim(-0.5, 1.5)
    plt.legend(loc = 'upper right')
    plt.show()

gaussian_function(0.5, 2, 3, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-4, 4, 100)

def gaussian_function(mu, sigma, x):
    a = (1 / (2 * math.pi * (sigma ** 2))) ** 1/2
    y = np.exp(-x**2)
    y1 = a * np.exp(-(x-mu)**2 / sigma**2)

    plt.figure(figsize=(4, 4))
    plt.plot(x, y, 'black', linewidth=3, label='$y=exp(-x^2)$')
    plt.plot(x, y1, 'red', linewidth=3, label='$y=0.5exp(-\frac{(x-2)^2}/{3^2})$')
    plt.grid(True)
    plt.xlim(-4, 4)
    plt.ylim(-0.5, 1.5)
    plt.legend(loc = 'upper right')
    plt.show()

gaussian_function(2, 3, x)
```

## 4.7.9 2차원 가우스 함수

가우스 함수를 2차원으로 확장할 수 있음. 2차원 가우스 함수는 9장 가우시안 모델에서 등장한다.

$Input : x = [x_0, x_1]^T$ 라고 했을 때, 가우스 함수의 기본형은 다음과 같다.

$y = exp{(-(x_0^2 + x_1^2)})$

이것은 다음과 같이 변경이 가능

$y = a\cdot exp{(-\frac{1}{2}(x-μ)^T𝚺^-1(x-μ)})$

먼저 함수의 형태를 나타내는 매개 변수는 μ와 𝚺이다. μ는 평균 벡터 (중심 벡터) 로 불리는 매개 변수이며, 분포의 중심을 나타낸다.

$μ = [μ_0, μ_1]^T$

------------------------------------------

𝚺는 공분산 행렬로, 2x2 행렬을 의미한다.

$𝚺=\begin{bmatrix}
σ_0, σ_2 \\
σ_2, σ_1
\end{bmatrix}$
------------------------------------------

행렬 요소의 $σ_0$와 $σ_1$, $x_0$ 방향과 $x_1$ 방향의 분포의 퍼짐을 나타낸다. $σ_2$는 분포의 기울기 ($x_0$과 $x_1$의 상관 관계)에 대응하고 있으며, 양수라면 오른쪽으로 치우친 분포가, 음수를 넣으면 왼쪽으로 치우친 분포가 표현된다.

exp의 내용은 벡터나 행렬이 들어가 있지만, 스칼라 양을 표현하는 것이다. 

$μ = [μ_0, μ_1]^T = [0, 0]^T$라고 했을 때, $(x-μ)^T𝚺^-1(x-μ)$를 계산하면,

$= \displaystyle [x_0, x_1]\cdot\frac{1}{σσ_1 - σ_2^2}\begin{bmatrix}
σ_1, -σ_2 \\
-σ_2, σ_0
\end{bmatrix}\begin{bmatrix}
x_0 \\
x_1
\end{bmatrix}$

= $\frac{1}{σσ_1 - σ_2^2}(σ_1x_0^2-2σ_2x_0x_1+σ_0x_1^2)$

α는 함수의 크기를 제어하는 매개 변수로 생각할 수도 있지만, 가우스 함수에서 확률 분포를 나타내는 경우는 다음과 같이 설정

$α = \frac{1}{2\pi}\frac{1}{|𝚺|^{1/2}}$

위처럼 하면 입력 공간에서 적분 값이 1이고, 함수가 확률 분포를 표현할 수 있음.

|𝚺|는 행렬식으로써, 2x2 행렬의 경우에는 다음과 같다.

$|A| = \begin{vmatrix}
a b \\
c d\end{vmatrix} = ad - bc$

따라서 |𝚺| = $σ_0σ_1 - σ_2^2$

```python
# list 4-5-1
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

# ガウス関数
def gauss(x, mu, sigma):
    N, D = x.shape
    c1 = 1 / (2 * np.pi)**(D / 2)
    c2 = 1 / (np.linalg.det(sigma)**(1 / 2))
    inv_sigma = np.linalg.inv(sigma)
    c3 = x - mu
    c4 = np.dot(c3, inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5 + c4[:, d] * c3[:, d]
    p = c1 * c2 * np.exp(-c5 / 2)
    return p
```

```python
# list 4-5-2
x = np.array([[1, 2], [2, 1], [3, 4]])
mu = np.array([1, 2])
sigma = np.array([[1, 0], [0, 1]])
print(gauss(x, mu, sigma))
```

```python
# list 4-5-3
X_range0=[-3, 3]
X_range1=[-3, 3]

# 等高線表示
def show_contour_gauss(mu, sig):
    xn = 40 # 等高線表示の解像度
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn, 1), np.reshape(xx1, xn * xn, 1)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    cont = plt.contour(xx0, xx1, f, 15, colors='k')
    plt.grid(True)
    
# 3D表示
def show3d_gauss(ax, mu, sig):
    xn = 40 # 等高線表示の解像度
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)    
    x = np.c_[np.reshape(xx0, xn * xn, 1), np.reshape(xx1, xn * xn, 1)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    ax.plot_surface(xx0, xx1, f, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')
    
# メイン
mu = np.array([1, 0.5])
sigma = np.array([[2, 1], [1, 1]])
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
